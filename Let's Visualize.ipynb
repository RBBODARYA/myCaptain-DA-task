```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Creating the score report

```python
Cycle= ["cycle 1","cycle 2","cycle 3","cycle 4"]
Score= [80, 75, 90, 94]
```

### Visualizing the score report

```python
plt.bar(Cycle,Score,color="blue")
plt.xlabel("Cycle")
plt.ylabel("Score")
plt.title("Score Report")
plt.ylim(0,100)
```
